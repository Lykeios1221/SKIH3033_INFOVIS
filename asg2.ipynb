```python
import ee
import geemap
import geojson
```

```python
ee.Initialize()
```

```python
with open("data/asg1/malaysia_states.geojson") as f:
    states = geojson.load(f)

geometry = ee.Geometry.MultiPolygon(states["features"][4]["geometry"]["coordinates"])
```

```python
startDate = "2024-01-01"
endDate = "2024-05-15"

s2 = (
    ee.ImageCollection("COPERNICUS/S2_HARMONIZED")
    .filterDate(startDate, endDate)
    .filterBounds(geometry)
    .filter(ee.Filter.lt("CLOUDY_PIXEL_PERCENTAGE", 35))
)
```

```python
s2Image = ee.Image(s2.first())
s2VisParams = dict(bands= ['B4', 'B3', 'B2'], min= 0, max= 3000)

Map = geemap.Map()
Map.addLayer(s2Image, s2VisParams, 'Sentinel-2 Image')
Map.centerObject(geometry, 13)

Map.save("map.html")
```

```python
imageId = s2Image.get('system:index')
print(imageId)
```
